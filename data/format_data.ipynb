```python
import pandas as pd

groups = pd.read_csv("./Groups.txt", usecols=["ecode", "pcode", "pabbrev"], index_col="pcode")
groups = groups[groups.ecode == 5]

candidates = pd.read_csv("./Candidates.txt", usecols=["ecode", "pcode", "cname"])
candidates = candidates[candidates.ecode == 5]
candidates["name_last"] = candidates.cname.apply(lambda x: x.split(", ")[0].strip())
candidates["name_first"] = candidates.cname.apply(lambda x: x.split(", ")[1].strip())
candidates["party"] = candidates.pcode.apply(lambda x: groups.at[x, "pabbrev"])
candidates = candidates[["name_last", "name_first", "party"]]
candidates.set_index("name_last", inplace=True)

data = pd.read_csv("./election_data.csv").T
votes = []
for name in data.index:
    votes.append(list(data.loc[name]))
data.index = list(candidates.index) + ["exhausted"]
data["name_first"] = list(candidates.name_first) + [""]
data["votes"] = votes
data["party"] = list(candidates.party) + [""]
data = data[["name_first", "party", "votes"]]
data.to_csv("./electionData.csv", index_label="name_last")
data
```
